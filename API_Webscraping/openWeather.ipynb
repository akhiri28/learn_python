```python
output = response.json()
output.keys()
```

```python
dt = []
aqi = []
co = []
no = []
no2 = []
o3 = []
so2 =[]
pm2_5 = []
pm10 = []
nh3 = []
```

```python
for otpt in output['list']:
    dt.append(otpt['dt'])
    aqi.append(otpt['main']['aqi'])
    co.append(otpt['components']['co'])
    no.append(otpt['components']['no'])
    no2.append(otpt['components']['no2'])
    o3.append(otpt['components']['o3'])
    so2.append(otpt['components']['so2'])
    pm2_5.append(otpt['components']['pm2_5'])
    pm10.append(otpt['components']['pm10'])
    nh3.append(otpt['components']['nh3'])
```

```python
air_pol_nagole_1yr_df = pd.DataFrame({'dt': dt, 'aqi' : aqi, 'co': co, 'no' : no,
              'no2': no2, 'o3' : o3, 'so2': so2, 'pm2_5' : pm2_5,
              'pm10': pm10, 'nh3': nh3})
```
